### 4.定义一个函数，完成一个小任务：对于给定的银行定期利率(输入)，计算多少年后可以连本带息翻番

```python
def helper(rates):
    n = 1
    while True:
        if (1 + rates) ** n >= 2:
            break
        n += 1
    print (n, "years")

helper(0.03)
```

### 5.一个数如果恰好等于它的因子之和，这个数就称为“完数”。例如，6的因子为1、2、3，而6=1+2+3，因此6是完数。编程，找出10000之内的所有完数，并输出该完数及对应的因子。

```python
def wanshu(N):
    factors = []
    for i in range(1, N):
        if N % i == 0:
            factors.append(i)   # find all factors
    if sum(factors) == N:
        return True, factors
    return False, []

for i in range(1, 10000):
    isW, facs = wanshu(i)
    if isW:
        print i, facs
```

### 6.摘录网页HTML源码，粘贴到input.txt，统计其中英文字母、空格、数字和其他字符的个数并输出。

```python
counts = {"letter": 0, "blank": 0, "number": 0, "other": 0}

f = open('./input_1.txt', 'r')
contents = f.readlines()
f.close()
for coni in contents:
    tmp = coni.strip()
    for ci in tmp:
        if ci.isalpha():    # is letter?
            counts["letter"] += 1
        elif ci.isspace():  # is space?
            counts["blank"] += 1
        elif ci.isdigit():  # is number?
            counts["number"] += 1
        else:               # other
            counts["other"] += 1
for item in counts.items():
    print (item[0] + ": " + str(item[1]))
```

### 7.在网上摘录一段英文文本(尽量长一些)，粘贴到input.txt，统计其中每个单词的词频(出现的次数)，并按照词频的顺序写入out.txt文件，每一行的内容为“单词:频次”

```python
import re

f = open('./input_2.txt', 'r')
contents = f.read()
f.close()
words = re.findall(r'\w+', contents)
counts = {}
for wordi in words:
    if wordi in counts:
        counts[wordi] += 1
    else:
        counts[wordi] = 1
tmp = sorted(counts.items(), key = lambda item:item[1], reverse = True)

fw = open('./out.txt', 'w')
for ti in tmp:
    fw.write(ti[0] + ": " + str(ti[1]) + "\n")
fw.close()
```

### 8. 王同学希望用电脑记录他每天掌握的英文单词。请设计程序和相应的数据结构，使小王能记录新学的英文单词和其中文翻译，并能很方便地根据英文来查找中文。实现一个类，能完成功能：1.背单词(添加单词)  2.查单词(根据单词进行翻译) 3.修改单词含义 4.统计历届四六级试卷，查找高频词，按照高频词排序

```python
import zipfile
import re

class wordMachine:
    def __init__(self):   # initialization
        self.dictionary = {}
        
    def addWord(self, EN, CN):  # add words
        if EN in self.dictionary:
            self.dictionary[EN].append(CN)  # 一词多义
        else:
            self.dictionary[EN] = [CN]
        
    def searchWord(self, EN):   # search words
        if EN in self.dictionary:
            return self.dictionary[EN]
        else:
            print "The word '" + EN + "' is not in dictionary!"
    
    def modifyWord(self, EN, CN): # modify words
        if EN in self.dictionary:
            self.dictionary[EN] = CN
        else:
            print "The word '" + EN + "' is not in dictionary!"
    
    def findTopKWords(self, K):  # get top K words in data
        azip = zipfile.ZipFile('./4 四六级英语试卷数据.zip')
        wordCount = {}
        for ni in azip.namelist()[1:]:
            contents = azip.read(ni)
            words = re.findall(r'\w+', contents)
            for wordi in words:
                if wordi in wordCount:
                    wordCount[wordi] += 1
                else:
                    wordCount[wordi] = 1
        tmp = sorted(wordCount.items(), key = lambda item:item[1], reverse = True)
        
        print "The top " + str(K) + " high frequency words are:" 
        for i in range(K):
            print tmp[i][0] + ": " + str(tmp[i][1])

# ----------------------  Test ----------------------------            
dics = wordMachine()
# ----- test add words
print "---------- 1. Test add words ----------"
dics.addWord("like", "喜欢")  
dics.addWord("send", "发送")
dics.addWord("like", "像")   # 一词可能有多义
print dics.dictionary
# ----- test search words
print "---------- 2. Test search words ----------"
dics.searchWord("like")  # exist
dics.searchWord("wo")  # not exist
# ----- test modify words
print "---------- 3. Test modify words ----------"
dics.modifyWord("send", "发出")
dics.modifyWord("wsd", "123")
print dics.dictionary
# ----- test get top K words in data
print "---------- 4. Test get top K words in data ----------"
dics.findTopKWords(10)
```

### 12.找出文件夹下所有文件名格式为output_YYYY.MM.DD.txt(output_2016.10.21.txt)的文件 。读取文件名中的日期时间信息，并找出这一天是周几。将文件改名为output_YYYY-MM-DD-W.txt (YYYY:四位的年，MM：两位的月份，DD：两位的日，W：一位的周几，并假设周一为一周第一天)

```python
import os
import re
import datetime
path = "./Q12 docs"
flist = os.listdir(path)
pattern = re.compile(r"^output_\d{4}\.\d+\.\d+\.txt$")
for fi in flist:
    m = re.search(pattern, fi)
    if m:
        year, month, day = m.group()[7:-4].split(".")
        datenum = datetime.date(int(year), int(month), int(day))
        newname = "output_" + datenum.strftime('%Y-%m-%d-%w') + ".txt"
        os.rename(path + "/" + m.group(), path + "/" + newname)
```
